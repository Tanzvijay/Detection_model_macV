```python
import cv2
import pandas as pd
from collections import defaultdict


# Load the YOLO model
model = YOLO("yolov8n.pt")

# Open the video file
video_path = "macv-obj-tracking-video (2).avi"
cap = cv2.VideoCapture(video_path)

# Prepare to track objects and times
object_times = defaultdict(lambda: {"start": None, "end": None})
unique_object_ids = set()
frame_rate = cap.get(cv2.CAP_PROP_FPS)
frame_count = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Perform object detection
    results = model(frame)

    # Extract the detections
    detections = results[0]  # Assuming results is a list of detections

    # Process each detected object
    for detection in detections.boxes:
        object_id = int(detection.cls)  # Get the object class or ID
        unique_object_ids.add(object_id)

        # Track the time the object appears
        if object_times[object_id]["start"] is None:
            object_times[object_id]["start"] = frame_count / frame_rate
        object_times[object_id]["end"] = frame_count / frame_rate

    frame_count += 1

cap.release()

# Prepare data for CSV
output_data = []
for obj_id, times in object_times.items():
    total_time = times["end"] - times["start"]
    output_data.append({"Object ID": obj_id, "Start Time (s)": times["start"],
                        "End Time (s)": times["end"], "Total Time (s)": total_time})

# Export to CSV
df = pd.DataFrame(output_data)
df.to_csv("object_tracking_results.csv", index=False)

# Print results
print(f"Total unique objects detected: {len(unique_object_ids)}")
print(f"Results saved to object_tracking_results.csv")
```
